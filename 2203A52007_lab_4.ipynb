```python
import pandas as pd
a=pd.read_csv("/content/Salary_Data.csv")
print(a)
```

```python
features = a["YearsExperience"]
print(features)
```

```python
target=a["Salary"]
print(target)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.xlabel("features")
plt.ylabel("target")
plt.title("Graph between features and target")
plt.plot(a['YearsExperience'],a['Salary'],label="YearsExperience")
plt.show()
#plt.scatter(features,target)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the salary dataset

# Create a linear regression object
lr = LinearRegression()

lr.fit(a[['YearsExperience']], a['Salary'])

# Print the slope and intercept of the line
print('Slope:', lr.coef_)
print('Intercept:', lr.intercept_)

# Plot the data points and the best fit line
plt.scatter(a['YearsExperience'], a['Salary'])
plt.plot(a['YearsExperience'], lr.predict(a[['YearsExperience']]), color='red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Best Fit Line')
plt.show()
```
